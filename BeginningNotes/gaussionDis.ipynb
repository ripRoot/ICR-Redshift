# What is Gaussian Distribution?

The Gaussian distribution is a bell-shaped curve that shows how values are spread out in many natural situations.



The equation looks like this:

$$ f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2} $$


$x$ represents the Variable

$\mu$ represents the Mean

$\sigma$ represents standard deviation

$e$ represents the base of the natural log.

properties of Gaussian Distribution:
    
    -Symmetric around it's mean
    -unimodal (one MODE) (means only one peak)
    -Asymptotic

# What is Standard Deviation ?

Standard deviation is the measure of spread.
Small: values are packed tightly around the mean
Big: Values are more spread out.

# What is One Standard Deviation?

In a gauissian distribution:
    
    -68% of the values lie within 1 standard deviation of the mean
    (This is what it means when people say something is within "1sd")
    -95% of values lie within 2sd
    -99.7 lie within 3sd

    



Standard deviations are used to subdivide the area under the normal curve.

# What does it mean to set a threshold>?

    -Threshold is a cut-off point where you choose to decide:
            Everything above this value is inportant. below is not.
    -EX in signal detection:
        I'm analyzing a graph of signal intensity.
        I set a threshold of 0.5.
        Now I say: only peaks above 0.5 are real. ignore the rest.
    

# What does it Mean to choose a distance?

    -In signal processing, distance often means:
        The space (in time, frequency, or index) between points.
        Used to avoid detecting the same peak multiple times

        EX:
            I detect a peak at wavelength = 10
            So I wait 5 steps before accepting another peak
            set a minimum distance between peaks = 5
