Some astronomical objects emit mostly infrared radiation, others mostly visable light, and still others mostly ultraviolet radiation. What determines this is their tempature.

The individual particcles vibrate in place or move rapidly from collisions, so the emitted waves are, on average, more energetic. Recall that higher energy waves have a higher frequency.

The relationship between tempature and electromagnetic radiation, we imagine an idealized object called a blackbody.

The graph shows the power emitted at each wavelength by objects of different tempatures.



Notice that the curves show that, at each tempature, our blackbody object emits radiation (photons) at all wavelengths (all colors). This is because in any soild or denser gas, some molecules or atoms vibrate or move between collisions slower than average and some move faster than average. 

In the taller curves in the graph above, the atoms have more collisions and give off more energy. In the real world of stars, this means that hotter stars give off mor eenergy at every wavelength than do cooler stars.


Hotter tempature = shorter wavelength (blue)
Cooler tempature = longer wavelength (red)

see pages 14 for more explaination.

There is a more precise star thermometer by measuring how much energy a star gives off at each wavelength. By constructing diagrams like the one about, the location of the maximum in the power curve of each star can tell us it's radiation. 

$$ \lambda_m = \frac{3\times10^6}{T} $$

This relationship is Wien's law. This is how we can calculate the wavelength at which maximum power is emitted. The wavelength is in nanometers, tempature is in K. 

For the Sun, the wavelength at which the maximum energy is emitted is 520 nanometers, which is near the middle of that portion of the electromagnetic spectrum called visable light.

We can also describe our observation that hotter objects radiate more power at all wavelengths in a mathematical form. If we sum up the contributions from all parts of the electromagnetic psectrum, we obtain the total energy emitted by a blackybody. What we usually measure from a large object like a star is the energy flux, the power emitted per square meter.

The energy flux from a blackbody at tempature T is proportional to the fourth power of it's absolute tempature. This relationship is the Stefan-Boltzmann law

$$ F = \sigma T^4 $$

Where F is the energy flux and $ \sigma $ is a constant number ($5.67\times 10^-8$)

Light can be reflected from a surface. With a very smooth and shiny surface, such as a mirror, the direction of the reflected light beam can be calculated accurately from knowledge of the shape of the reflecting surfacce. *Large telescopes depend entirely on mirrors for their main opictal elements.

in a prisim, the path of light is refracted (bent) but not all of the colors are bent by the same ammount. The bending relies on the wavelength of the light as well as the properties of the material. As a result, different wavelengths are bent by different amounts which follow different paths through the prisim. Violet light is bent more than red.

This is called dispersion and explains Newton's rainbow experiment.



if the prisim is focused on a screen, then the different wavelengths (or colors) that make up white light are lined up side by side like a rainbow.

The instrument used to disperse the light and form the spectrum is called a spectrometer.



In 1802, William Wollaston built an improved spectrometer that included a lens to focus the Sun's spectrum on a screen. He saw that the colors were not spready uniformly, but instead some of the ranges of color were missing. It appearing as dark bands in the spectrum. Upon a more careful examination of the solar spectrum, he found 600 such dark lines.

Researchers had found that gases turned out to not be transparent at all colors, but quite opaque at a few sharply defined wavelengths. Different elements absorbed different colors and such that certain lines in the spectrum "go with" certain elements. Like a fingerprint.

Puree Hydrogen would emit one pattern of color, pure sodium would emit a different pattern. A mixture of hydrogen and sodium emitted both sets of spectral lines. 

Types of spectra:

    Continuous Spectrum: Formed when a solid or very dense gas gives off radiation, is an array of all wavelengths or colors of the rainbow.

    Absorption Spectrum: consists of a series or pattern of dark lines, missing colors, superimposed upon the continous spectrum of a source.

    Emission Spectrum: Appears as a pattern or series of bright lines; it consists of light in which only certain discrete wavelengths are present. 

In contrast to the fingerprint idea above, Absorption spectra occur when passing white light through a cool, thin gas. Tempature and other conditions determine whether th elines are bright or dark, but thee wavelengths of the lines for any element are the same in either case.

Spectral analysis is quite useful because it can be applied to light reflected off the surface of a nearby asteroid as well as light from a distant galaxy. 

**The dark lines in the solar spectrum give us evidence that there are certain chemicals between us and the sun.

We can use the presence of absorption and emission lines to analyze the composition of other stars and clouds of gas in space.

Rutherford's model had a problem: according to electromagnetic theory, orbiting electrons should constantly emit energy and quickly spiral into the nucleus. Niels Bohr solved this by proposing that electrons can only exist in certain orbits.

Bohr suggested that electrons radiate no energy as long as they move within an allowed orbit. Energy changed only occur when an electron moves from one orbit to another.

in a Bohr model, each electron orbit has a specific energy value (energy level). An electron moving from a higher orbit to a lower orbit released energy in the form of electromagnetic radiation, and vise versa.

Think of a building and as you go up the price increases. Electrons can only occupy specific energy levels, and energy is required to move at a higher level.

When electrons transition between energy levels, they emit or absorb packets of energy called photons. The energy (E) of a photon is related to frequency (f) by the formula:

$$ E = hf $$

h is Planck's constant ($6.626\times10^-34$)
